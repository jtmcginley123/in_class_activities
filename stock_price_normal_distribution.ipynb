```python
import os
import pandas as pd
import alpaca_trade_api as tradeapi 

from dotenv import load_dotenv
load_dotenv()

%matplotlib inline
```

```python
alpaca_api_key = os.getenv("ALPACA_API_KEY") # you would need to create a .env and a .gitignore and put your keys in there and reference them here and 
alpaca_secret_key = os.getenv("ALPACA_SECRET_KEY") # here as well

alpaca = tradeapi.REST(
    alpaca_api_key,
    alpaca_secret_key,
    api_version="v2"
)
```

```python
ticker = ["TSLA", "KO"]

timeframe = "1D"

start_date = pd.Timestamp("2019-05-01", tz="America/New_York").isoformat()
end_date = pd.Timestamp("2020-05-01", tz="America/New_York").isoformat()

df_ticker = alpaca.get_barset(
    ticker,
    timeframe,
    start=start_date,
    end=end_date
).df

df_ticker.head(10)
```

```python
df_closing_prices = pd.DataFrame()


df_closing_prices["KO"] = df_ticker["KO"]["close"]
df_closing_prices["TSLA"] = df_ticker["TSLA"]["close"]

df_closing_prices.index = df_closing_prices.index.date
```

```python
df_daily_returns = df_closing_prices.pct_change().dropna()

df_daily_returns.head(10)
```
